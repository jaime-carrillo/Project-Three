```python
# Read in 2019 file
df_2019 = pd.read_csv('..//..//data/hospitals/2019-ed-pivot-table_20201013.csv')

# Fill in all blanks with 0 (for deep learning)
df_2019.fillna(0, inplace=True)
df_2019.dropna(inplace=True)
```

```python
# Read in 2018 file
df_2018 = pd.read_csv('..//..//data/hospitals/2018-ed-pivot-table.csv')

# Fill in all blanks with 0 (for deep learning)
df_2018.fillna(0, inplace=True)
df_2018.dropna(inplace=True)
```

```python
# Read in 2017 file
df_2017 = pd.read_csv('..//..//data/hospitals/2017-ed-pivot-table20180718.csv')

# Fill in all blanks with 0 (for deep learning)
df_2017.fillna(0, inplace=True)
df_2017.dropna(inplace=True)
```

```python
# Read in 2016 file
df_2016 = pd.read_csv('..//..//data/hospitals/2016-ed-pivot-table_20181210.csv')

# Fill in all blanks with 0 (for deep learning)
df_2016.fillna(0, inplace=True)
df_2016.dropna(inplace=True)
```

```python
# Read in 2015 file
df_2015 = pd.read_csv('..//..//data/hospitals/2015-emergency-department-pivot.csv')

# Fill in all blanks with 0 (for deep learning)
df_2015.fillna(0, inplace=True)
df_2015.dropna(inplace=True)
```

```python
# Merge all files
all_df = df_2019.append([df_2018, df_2017, df_2016, df_2015], ignore_index=True)
df_1 = all_df.dropna(axis='columns')
df_1
```

```python
# Merge last 3 years files
pd.set_option('max_columns', None)
threeyear_df = df_2019.append([df_2018, df_2017], ignore_index=True)
df_3 = threeyear_df.dropna(axis='columns')
df_3
```
